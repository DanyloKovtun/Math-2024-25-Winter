# Wektory  - Rzuty i podziały

## Rzutowanie wektora na wektor

Rzutowanie wektora $\mathbf{a}$ na oś $\mathbf{b}$ polega na znalezieniu składowej wektora $\mathbf{a}$ wzdłuż kierunku $\mathbf{b}$.

Rzut $\mathbf{a}$ na $\mathbf{b}$:

$$
\text{Proj}_{\mathbf{b}} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{b}}{|\mathbf{b}|^2} \mathbf{b}
$$


## Rzutowanie wektora na osie

Rzutowanie wektora $\mathbf{a}$ na oś polega na znalezieniu składowej wektora $\mathbf{a}$ wzdłuż osi.

Rzut $\mathbf{a}$ na oś $OX$:

$$
\text{Proj}_{OX} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{i}}{|\mathbf{i}|^2} \mathbf{i} = \frac{\mathbf{a} \cdot \mathbf{i}}{1} \mathbf{i} = (\mathbf{a} \cdot \mathbf{i}) \mathbf{i}
$$

Rzut $\mathbf{a}$ na oś $OY$:

$$
\text{Proj}_{OY} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{j}}{|\mathbf{j}|^2} \mathbf{j} = \frac{\mathbf{a} \cdot \mathbf{j}}{1} \mathbf{j} = (\mathbf{a} \cdot \mathbf{j}) \mathbf{j}
$$

Przypominamy, że $\mathbf{i} = (1, 0)$ i $\mathbf{j} = (0, 1)$ są wektorami jednostkowymi wzdłuż osi $OX$ i $OY$.


## Dzielenie odcinka w określonym stosunku

Odcinek jest zdefiniowany przez dwa punkty w przestrzeni.

Odcinki i wektory są ściśle powiązane, ponieważ odcinek łączący dwa punkty można traktować jako wektor skierowany od jednego punktu do drugiego.

Jeśli punkt \( P \) dzieli odcinek łączący punkty $A(x_1, y_1)$ i $B(x_2, y_2)$ w stosunku $m:n$ to współrzędne punktu $P$ można znaleźć z wzoru:

$$
P = \left(\frac{n \times x_2 + m \times x_1}{m+n}, \frac{n \times y_2 + m \times y_1}{m+n}\right)
$$

**Przykład:**

Dany jest odcinek łączący punkty $A(1, 2)$ i $B(3, 8)$. Znajdź punkt $P$, który dzieli odcinek w stosunku 2:3.

Współrzędne punktu $P$:
$$
P = \left(\frac{2 \times 3 + 3 \times 1}{2+3}, \frac{2 \times 8 + 3 \times 2}{2+3}\right) = \left(\frac{9}{5}, \frac{22}{5}\right)
$$

```python
import matplotlib.pyplot as plt
import math

# Współrzędne punktów A i B
A = (1, 2)
B = (3, 8)

# Stosunek m:n
m, n = 2, 3

# Obliczenie współrzędnych punktu P, który dzieli odcinek w stosunku 2:3
P_x = (n * A[0] + m * B[0]) / (m + n) # błąd, musi być (n * B[0] + m * A[0]) / (m + n)
P_y = (n * A[1] + m * B[1]) / (m + n) # błąd, musi być (n * B[1] + m * A[1]) / (m + n)
P = (P_x, P_y)

# Rysowanie wykresu
plt.figure(figsize=(6, 6))

# Rysowanie odcinka AB
plt.plot([A[0], B[0]], [A[1], B[1]], 'k--', label='Odcinek AB')

# Zaznaczanie punktów A, B i P
plt.scatter(*A, color='blue', label='A(1, 2)')
plt.scatter(*B, color='red', label='B(3, 8)')
plt.scatter(*P, color='green', label=f'P({P[0]:.2f}, {P[1]:.2f})')

# Opisy punktów
plt.text(A[0]-0.2, A[1], 'A(1, 2)', fontsize=12, color='blue')
plt.text(B[0]+0.1, B[1], 'B(3, 8)', fontsize=12, color='red')
plt.text(P[0]+0.1, P[1], f'P({P[0]:.2f}, {P[1]:.2f})', fontsize=12, color='green')

# Dodanie siatki i etykiet osi
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Dzielenie odcinka w stosunku 2:3')
plt.legend()

# Wyświetlenie wykresu
plt.show()
```

- Wykonaj rzutowanie wektora $\mathbf{b} = (1, 1)$ na wektor $\mathbf{a} = (2, 3)$. Z ilustruj wynik na wykresie.

```python
# 1

import matplotlib.pyplot as plt
import math

# Współrzędne punktów A i B (Координаты точек A и B)
A = (-1, 2)
B = (3, -2)

# Stosunek m:n (Соотношение m:n)
m, n = 1, 3

# Obliczenie współrzędnych punktu P, który dzieli odcinek w stosunku m:n
# (Вычисление координат точки P, которая делит отрезок в отношении m:n)
P_x = (n * B[0] + m * A[0]) / (m + n)  # Współrzędna x punktu P (Координата x точки P)
P_y = (n * B[1] + m * A[1]) / (m + n)  # Współrzędna y punktu P (Координата y точки P)
P = (P_x, P_y)  # Punkt P (Точка P)

# Rysowanie wykresu (Построение графика)
plt.figure(figsize=(6, 6))  # Tworzymy obszar wykresu o wymiarach 6x6 cali (Создаем область графика размером 6x6 дюймов)

# Rysowanie odcinka AB (Построение отрезка AB)
plt.plot([A[0], B[0]], [A[1], B[1]], 'k--', label='Odcinek AB')
# Funkcja plt.plot rysuje przerywaną linię ('k--') łączącą punkty A i B (Функция plt.plot рисует пунктирную линию ('k--'), соединяющую точки A и B)

# Zaznaczanie punktów A, B i P (Выделение точек A, B и P)
plt.scatter(*A, color='blue', label='A(-1, 2)')  # Rysowanie punktu A w kolorze niebieskim (Рисование точки A синим цветом)
plt.scatter(*B, color='red', label='B(3, -2)')  # Rysowanie punktu B w kolorze czerwonym (Рисование точки B красным цветом)
plt.scatter(*P, color='green', label=f'P({P[0]:.2f}, {P[1]:.2f})')  # Rysowanie punktu P w kolorze zielonym (Рисование точки P зеленым цветом)

# Opisy punktów (Подписи точек)
plt.text(A[0]-0.2, A[1], 'A(-1, 2)', fontsize=12, color='blue')  # Tekst z etykietą punktu A (Подпись точки A)
plt.text(B[0]+0.1, B[1], 'B(3, -2)', fontsize=12, color='red')  # Tekst z etykietą punktu B (Подпись точки B)
plt.text(P[0]+0.1, P[1], f'P({P[0]:.2f}, {P[1]:.2f})', fontsize=12, color='green')  # Tekst z etykietą punktu P (Подпись точки P)

# Dodanie siatki i etykiet osi (Добавление сетки и подписей к осям)
plt.grid(True)  # Wyświetlenie siatki na wykresie (Отображение сетки на графике)
plt.xlabel('x')  # Etykieta osi X (Подпись оси X)
plt.ylabel('y')  # Etykieta osi Y (Подпись оси Y)
plt.title('Dzielenie odcinka w stosunku 1:3')  # Tytuł wykresu (Название графика)
plt.legend()  # Wyświetlenie legendy (Отображение легенды)

# Wyświetlenie wykresu (Показ графика)
plt.show()  # Wyświetla wykres w oknie (Отображает график в окне)
```

```python
# 2

import numpy as np
import matplotlib.pyplot as plt

# Definiujemy wektory jako obiekty tablic (Определяем векторы как массивы)
a = np.array([3, 4])
i = np.array([1, 0])
j = np.array([0, 1])

# Obliczamy iloczyn skalarny wektora a i i (Находим скалярное произведение вектора a и i)
dotAI = a.dot(i) # ((3, 4) * (1, 0)) = (3 * 1 + 4 * 0) = 3
proj_OX_a = dotAI * i  # Rzut wektora a na oś OX (Проекция вектора a на ось OX) 3 * (1, 0) = (3, 0)

# Obliczamy iloczyn skalarny wektora a i j (Находим скалярное произведение вектора a и j)
dotAJ = a.dot(j) # ((3, 4) * (0, 1)) = (3 * 0 + 4 * 1) = 4
proj_OY_a = dotAJ * j  # Rzut wektora a na oś OY (Проекция вектора a на ось OY) 4 * (0, 1) = (0, 4)

# Ustawienia wykresu (Настройки графика)
plt.figure(figsize=(6, 6))
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)  # Dodajemy oś poziomą (Добавляем горизонтальную ось)
plt.axvline(0, color='black', linewidth=0.5)  # Dodajemy oś pionową (Добавляем вертикальную ось)

# Rysujemy wektor a (Рисуем вектор a)
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Wektor a')

# Rysujemy projekcje na osie OX i OY (Рисуем проекции на оси OX и OY)
plt.quiver(0, 0, proj_OX_a[0], proj_OX_a[1], angles='xy', scale_units='xy', scale=1, color='green', label='Rzut na OX')
plt.quiver(0, 0, proj_OY_a[0], proj_OY_a[1], angles='xy', scale_units='xy', scale=1, color='red', label='Rzut na OY')

# Ustawienia wykresu (Настройки графика)
plt.xlim(-1, 5)  # Zakres osi x (Диапазон оси x)
plt.ylim(-1, 5)  # Zakres osi y (Диапазон оси y)
plt.xlabel('OX')  # Etykieta osi OX (Подпись оси OX)
plt.ylabel('OY')  # Etykieta osi OY (Подпись оси OY)
plt.legend()  # Dodajemy legendę (Добавляем легенду)
plt.title('Rzut wektora a na osie OX i OY')  # Tytuł wykresu (Заголовок графика)
plt.show()  # Wyświetlamy wykres (Отображаем график)
```

```python
# 3

import numpy as np
import matplotlib.pyplot as plt

# Definiujemy wektory a i b (Определяем векторы a и b)
a = np.array([2, 3])
b = np.array([1, 1])

# Projab = ((a * b) / |b|^2) * b
# Obliczamy iloczyn skalarny wektora a i b (Вычисляем скалярное произведение векторов a и b)
dot_a = a.dot(b) # a * b

# Obliczamy iloczyn skalarny wektora b z samym sobą, co odpowiada jego długości podniesionej do kwadratu (Вычисляем скалярное произведение вектора b с самим собой, что соответствует его длине в квадрате)
dot_b = b.dot(b) # |b|^2

# Obliczamy współczynnik dla wektora b, który określa długość rzutu (Вычисляем коэффициент для вектора b, который определяет длину проекции)
p = dot_a / dot_b # (a * b) / |b|^2

# Obliczamy rzut wektora a na wektor b (Вычисляем проекцию вектора a на вектор b)
proj = p * b # ((a * b) / |b|^2) * b

# Konfiguracja wykresu (Настройка графика)
plt.figure(figsize=(6, 6))
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)  # Oś X (Ось X)
plt.axvline(0, color='black', linewidth=0.5)  # Oś Y (Ось Y)

# Rysujemy wektor a (Рисуем вектор a)
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Wektor a (2, 3)')

# Rysujemy wektor b (Рисуем вектор b)
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='purple', label='Wektor b (1, 1)')

# Rysujemy rzut wektora a na b (Рисуем проекцию вектора a на b)
plt.quiver(0, 0, proj[0], proj[1], angles='xy', scale_units='xy', scale=1, color='green', label=f'Proj_b a ({proj[0]:.2f}, {proj[1]:.2f})')

# Ustawienia osi i legendy (Настройки осей и легенды)
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.xlabel('Oś X')  # Podpis osi X (Подпись оси X)
plt.ylabel('Oś Y')  # Podpis osi Y (Подпись оси Y)
plt.legend()  # Dodanie legendy (Добавление легенды)
plt.title('Projekcja wektora a na wektor b')  # Tytuł wykresu (Название графика)
plt.show()  # Wyświetlenie wykresu (Отображение графика)
```

```python
# 4

import numpy as np
import matplotlib.pyplot as plt

# Definiujemy wektory a i b (Определяем векторы a и b)
a = np.array([2, 3])
b = np.array([1, 1])

# Obliczamy iloczyn skalarny wektora b z a (Вычисляем скалярное произведение вектора b с a)
dot_b = b.dot(a)

# Obliczamy iloczyn skalarny wektora a z samym sobą (Вычисляем скалярное произведение вектора a с самим собой)
dot_a = a.dot(a)

# Obliczamy współczynnik dla wektora a, który określa długość rzutu wektora b na a (Вычисляем коэффициент для вектора a, который определяет длину проекции вектора b на a)
p = dot_b / dot_a

# Obliczamy rzut wektora b na wektor a (Вычисляем проекцию вектора b на вектор a)
proj = p * a

# Konfiguracja wykresu (Настройка графика)
plt.figure(figsize=(6, 6))
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)  # Oś X (Ось X)
plt.axvline(0, color='black', linewidth=0.5)  # Oś Y (Ось Y)

# Rysujemy wektor a (Рисуем вектор a)
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Wektor a (2, 3)')

# Rysujemy wektor b (Рисуем вектор b)
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='purple', label='Wektor b (1, 1)')

# Rysujemy rzut wektora b na a (Рисуем проекцию вектора b на a)
plt.quiver(0, 0, proj[0], proj[1], angles='xy', scale_units='xy', scale=1, color='green', label=f'Proj_b a ({proj[0]:.2f}, {proj[1]:.2f})')

# Ustawienia osi i legendy (Настройки осей и легенды)
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.xlabel('Oś X')  # Podpis osi X (Подпись оси X)
plt.ylabel('Oś Y')  # Podpis osi Y (Подпись оси Y)
plt.legend()  # Dodanie legendy (Добавление легенды)
plt.title('Projekcja wektora a na wektor b')  # Tytuł wykresu (Название графика)
plt.show()  # Wyświetlenie wykresu (Отображение графика)
```
